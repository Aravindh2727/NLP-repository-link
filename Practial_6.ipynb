```python
import nltk
from nltk.tokenize import word_tokenize
from nltk import pos_tag

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('punkt_tab') # Added to fix the LookupError
nltk.download('averaged_perceptron_tagger_eng') # Added to fix specific tagger error
```

```python
sentence = "Natural Language Processing Help Computer Understand Human"

tokens = word_tokenize(sentence)

tags = pos_tag(tokens)
print("POS Tags:", tags)
```
